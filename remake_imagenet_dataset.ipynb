```python
import pandas as pd
```

```python
annot_path = "../data/tiny-imagenet-200/val/val_annotations.txt"

df = pd.read_csv(annot_path, sep='\t', header=None)
df = df.rename(columns={0: 'filename', 1:'foldername', 2:'a', 3:'b', 4:'c', 5:'d'})
```

```python
df.head()
```

```python
base_scr_path = "../data/tiny-imagenet-200/val/images/"
base_dst_path = "../data/tiny-imagenet-200/val_again/"

image_path = df.filename.to_list()

folder_pth = df.foldername.to_list()
```

```python
from shutil import copyfile
import os

made = []
for count in range(len(image_path)):
    
    cur_img = image_path[count]
    cur_fol = folder_pth[count]
    
    if cur_fol not in made:
        os.mkdir(base_dst_path + cur_fol)
        made.append(cur_fol)
        
    
    copyfile(base_scr_path+cur_img, base_dst_path+cur_fol+'/'+cur_img)
    
```
